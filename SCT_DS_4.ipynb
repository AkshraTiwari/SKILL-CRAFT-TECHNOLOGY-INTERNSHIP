```python
import pandas as pd
import numpy as np
import random
import datetime

# Set seed for reproducibility
random.seed(42)
np.random.seed(42)

# Number of accidents
num_accidents = 1000

# Generate synthetic data for accidents
data = {
    'Reference_Number': [f'ACC{i}' for i in range(1, num_accidents + 1)],
    'Grid_Ref_Easting': np.random.randint(50000, 70000, size=num_accidents),
    'Grid_Ref_Northing': np.random.randint(40000, 60000, size=num_accidents),
    'Number_of_Vehicles': np.random.randint(1, 5, size=num_accidents),
    'Number_of_Casualties': np.random.randint(1, 3, size=num_accidents),
    'Accident_Date': [datetime.date(2023, np.random.randint(1, 13), np.random.randint(1, 29)) for _ in range(num_accidents)],
    'Time_24hr': [f'{np.random.randint(0, 24):02d}:{np.random.randint(0, 60):02d}' for _ in range(num_accidents)],
    'First_Road_Class': [random.choice(['A', 'B', 'C']) for _ in range(num_accidents)],
    'Road_Surface': [random.choice(['Dry', 'Wet', 'Snow', 'Ice']) for _ in range(num_accidents)],
    'Lighting_Conditions': [random.choice(['Daylight', 'Darkness - lights lit', 'Darkness - lights unlit']) for _ in range(num_accidents)],
    'Weather_Conditions': [random.choice(['Fine', 'Raining', 'Snowing', 'Fog', 'Other']) for _ in range(num_accidents)],
    'Casualty_Class': [random.choice(['Driver', 'Passenger', 'Pedestrian']) for _ in range(num_accidents)],
    'Casualty_Severity': [random.choice(['Slight', 'Serious', 'Fatal']) for _ in range(num_accidents)],
    'Sex_of_Casualty': [random.choice(['Male', 'Female']) for _ in range(num_accidents)],
    'Age_of_Casualty': np.random.randint(18, 80, size=num_accidents),
    'Type_of_Vehicle': [random.choice(['Car', 'Van', 'Truck', 'Motorcycle']) for _ in range(num_accidents)]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the first few rows to verify
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting accident hotspots based on Grid Ref: Easting and Grid Ref: Northing
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Grid_Ref_Easting', y='Grid_Ref_Northing', data=df, hue='Road_Surface', palette='viridis', s=50)
plt.title('Accident Hotspots by Road Surface')
plt.xlabel('Grid Ref: Easting')
plt.ylabel('Grid Ref: Northing')
plt.legend(title='Road Surface')
plt.show()
```

```python
# Visualize accidents by weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Conditions', data=df, palette='Set2', order=df['Weather_Conditions'].value_counts().index)
plt.title('Weather Conditions causing Accidents')
plt.xticks(rotation=45)
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.show()

# Visualize accidents by time of day
plt.figure(figsize=(10, 6))
sns.countplot(x='Time_24hr', data=df, palette='Set2')
plt.title('Accidents by Time of Day')
plt.xlabel('Time (24hr)')
plt.ylabel('Number of Accidents')
plt.show()
```
